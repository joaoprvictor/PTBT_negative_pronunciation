```python
# create contingency table
data_crosstab_school = pd.crosstab(all_negatives['negation'],
                            all_negatives['schooling'], normalize= True,
                           margins=True, margins_name="Total").round(4)*100
data_crosstab_school
```

```python
# create contingency table
data_crosstab_age = pd.crosstab(all_negatives['negation'],
                            all_negatives['age'], normalize=True,
                           margins=True, margins_name="Total").round(4)*100

data_crosstab_age
```

```python
# create contingency table
data_crosstab_sex = pd.crosstab(all_negatives['negation'],
                            all_negatives['sex'], normalize=True,
                           margins=True, margins_name="Total").round(4)*100

data_crosstab_sex
```

```python
# significance level
alpha = 0.05

# Calcualtion of Chisquare
chi_square_school = 0
chi_square_age = 0
chi_square_sex = 0
```

```python
# chi2 schooling
rows_school = all_negatives['negation'].unique()
columns_school = all_negatives['schooling'].unique()
for i in columns_school:
    for j in rows_school:
        O = data_crosstab_school[i][j]
        E = data_crosstab_school[i]['Total'] * data_crosstab_school['Total'][j] / data_crosstab_school['Total']['Total']
        chi_square_school += (O-E)**2/E
```

```python
# chi2 age
rows_age = all_negatives['negation'].unique()
columns_age = all_negatives['age'].unique()
for i in columns_age:
    for j in rows_age:
        O = data_crosstab_age[i][j]
        E = data_crosstab_age[i]['Total'] * data_crosstab_age['Total'][j] / data_crosstab_age['Total']['Total']
        chi_square_age += (O-E)**2/E
```

```python
# chi2 sex
rows_sex = all_negatives['negation'].unique()
columns_sex = all_negatives['sex'].unique()
for i in columns_sex:
    for j in rows_sex:
        O = data_crosstab_sex[i][j]
        E = data_crosstab_sex[i]['Total'] * data_crosstab_sex['Total'][j] / data_crosstab_sex['Total']['Total']
        chi_square_sex += (O-E)**2/E
```

```python
# The p-value approach - school
import scipy.stats as stats
print("Approach 1: The p-value approach to hypothesis testing in the decision rule")
p_value = 1 - stats.chi2.cdf(chi_square_school, (len(rows_school)-1)*(len(columns_school)-1))
conclusion = "Failed to reject the null hypothesis."
if p_value <= alpha:
    conclusion = "Null Hypothesis is rejected. So the social variable can influence the negative pronunciation"
print("chisquare-score is:", chi_square_school, " and p value is:", p_value)
print(conclusion)
```

```python
# The p-value approach - age
import scipy.stats as stats
print("Approach 1: The p-value approach to hypothesis testing in the decision rule")
p_value = 1 - stats.chi2.cdf(chi_square_age, (len(rows_age)-1)*(len(columns_age)-1))
conclusion = "Failed to reject the null hypothesis."
if p_value <= alpha:
    conclusion = "Null Hypothesis is rejected. So the social variable can influence the negative pronunciation"
print("chisquare-score is:", chi_square_age, " and p value is:", p_value)
print(conclusion)
```

```python
# The p-value approach - sex
import scipy.stats as stats
print("Approach 1: The p-value approach to hypothesis testing in the decision rule")
p_value = 1 - stats.chi2.cdf(chi_square_sex, (len(rows_sex)-1)*(len(columns_sex)-1))
conclusion = "Failed to reject the null hypothesis."
if p_value <= alpha:
    conclusion = "Null Hypothesis is rejected. So the social variable can influence the negative pronunciation"
print("chisquare-score is:", chi_square_sex, " and p value is:", p_value)
print(conclusion)
```
